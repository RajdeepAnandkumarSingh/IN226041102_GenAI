Task 5: Input Validation Layer

```python
# Task 5: Input Validation Layer
def validate_inputs(audience, tone):
    # Allowed values [cite: 89]
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Assign default values if invalid [cite: 92]
    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "formal"

    return final_audience, final_tone

# Test validation
print(validate_inputs("expert", "fun"))
print(validate_inputs("alien", "angry")) # Should return defaults
```

Task 6: Prompt Generator App

```python
# Task 6: Prompt Generator App

def generate_prompt(topic, audience, tone, style):
    v_audience, v_tone = validate_inputs(audience, tone)

    styles = {
        "teaching": PromptTemplate.from_template(
            "Explain {topic} clearly step by step"
        ),
        "interview": PromptTemplate.from_template(
            "Ask 3 questions about {topic}"
        ),
        "storytelling": PromptTemplate.from_template(
            "Explain {topic} as a story"
        )
    }

    selected_template = styles.get(style, styles["teaching"])

    final_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template.template + " for {audience} in a {tone} style"
    )

    return final_template.format(
        topic=topic,
        audience=v_audience,
        tone=v_tone
    )

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

Task 7: Template Reusability Test

```python
# Task 7: Template Reusability Test

reusable_template = PromptTemplate.from_template(
    "Summarize {topic} for {audience} in a {tone} tone"
)

test_inputs = [
    {"topic": "Blockchain", "audience": "beginner", "tone": "casual"},
    {"topic": "Kubernetes", "audience": "expert", "tone": "formal"},
    {"topic": "Photosynthesis", "audience": "beginner", "tone": "fun"},
    {"topic": "Stock Market", "audience": "intermediate", "tone": "formal"},
    {"topic": "Cooking", "audience": "beginner", "tone": "casual"}
]

for item in test_inputs:
    v_audience, v_tone = validate_inputs(item["audience"], item["tone"])
    print(reusable_template.format(
        topic=item["topic"],
        audience=v_audience,
        tone=v_tone
    ))
```
